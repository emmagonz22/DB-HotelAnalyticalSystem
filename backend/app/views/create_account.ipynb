```python
from IPython.display import display, HTML
import ipywidgets as widgets
import ipyvuetify as v
import json
import requests
import api.entityCRUD as crud
```

```python
class Register:

    def __init__(self):
        
        self.out = widgets.Output()

        self.username = widgets.Text(
            placeholder='username'
        )

        self.password = widgets.Text(
            placeholder='password'
        )

        self.fname = widgets.Text(
            placeholder='fname'
        )

        self.lname = widgets.Text(
            placeholder='lname'
        )

        self.salary = widgets.Text(
            placeholder='salary'
        )

        self.age = widgets.Text(
            placeholder='age'
        )

        self.login_button = widgets.Button(
            description='Go to Login page',
        )
        self.submit_button = widgets.Button(
            description='Submit',
        )
     
        self.dropdown_options = ['Supervisor', 'Regular', 'Administrator']

        # Create the dropdown widget
        self.position_dropdown = widgets.Dropdown(options=self.dropdown_options, description='Select Option:', value='Regular')
     
    def hide_register(self):
        self.container.close()
        self.out.close()
        '''
        self.username.close()
        self.password.close()
        self.fname.close()
        self.lname.close()
        self.position.close()
        self.salary.close()
        self.age.close()
        self.submit_button.close()
        self.register_button.close()
        '''


    def register(self, b):
   

        # Validate data
        try:
            data_login =  {
                 'username': self.username.value,
                 'password': self.password.value
                 } 
            data_employee = {
                 'fname' : self.fname.value,
                 'lname' : self.lname.value,
                 'position' : self.position_dropdown.value,
                 'salary' : float(self.salary.value),
                 'age' : int(self.age.value)
                 } 

            #if isinstance(self.username.value, str) or isinstance(self.password.value, str):
            #    raise Exception("Username or password not a string")
            if not isinstance(data_employee['age'], int):
                raise Exception("Invalid parameters")
            if not isinstance(data_employee['salary'], float):
                raise Exception("Invalid parameters")

            # Create Login
            crud.createLogin(data_login)
            # Create Login
            crud.createEmployee(data_employee)
            
        except Exception as e:
            display(str(e))


        # Redirect to login 
        with self.out:
            self.out.clear_output()
            self.hide_register()
            %run login.ipynb

    def navigate_login(self, b):
        with self.out:
            self.out.clear_output()
            self.hide_register()
            %run login.ipynb

    def create_register(self):

        self.submit_button.on_click(self.register)
        self.login_button.on_click(self.navigate_login)


        self.container = widgets.VBox([self.username,
            self.password,
            self.fname,
            self.lname,
            self.position_dropdown,
            self.salary,
            self.age,
            self.login_button,
            self.submit_button]) 


        display(self.container, self.out)


```

```python
if __name__ == '__main__':
    register = Register()
    register.create_register()
```
