```python
import api.entityCRUD as entityCRUD
import matplotlib.pyplot as plt
import ipywidgets as widgets
import json
import os
```

```python
def create(messages):
    # Define the form elements
    rid = widgets.Text(description='Room Id:')
    startdate = widgets.Text(description='Start Date:')
    enddate = widgets.Text(description='End Date:')

    rid.style.description_width = 'initial'
    startdate.style.description_width = 'initial'
    enddate.style.description_width = 'initial'

    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not rid.value or not startdate.value or not enddate.value:
                raise Exception("Missing Data!")
            values = {
                'rid': int(rid.value),
                'startdate': startdate.value,
                'enddate': enddate.value,
            }
            return values
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([rid, startdate, enddate])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Create')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.createRoomUnavailable(json_data).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Room Unavaliable created with id: {data['ruid']}")
            except Exception as e:
                display(str(e))


    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
def update(messages):
    # Define the form elements
    ruid = widgets.Text(description='Room Unavailable Id:')
    rid = widgets.Text(description='Room Id:')
    startdate = widgets.Text(description='Start Date:')
    enddate = widgets.Text(description='End Date:')

    ruid.style.description_width = 'initial'
    rid.style.description_width = 'initial'
    startdate.style.description_width = 'initial'
    enddate.style.description_width = 'initial'

    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not ruid.value or not rid.value or not startdate.value or not enddate.value:
                raise Exception("Missing Data!")
            values = {
                'ruid': int(ruid.value),
                'rid': int(rid.value),
                'startdate': startdate.value,
                'enddate': enddate.value,
            }
            return values
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([ruid, rid, startdate, enddate])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Update')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.updateRoomUnavailable(json_data, ruid.value).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Room Unavailable {data['ruid']} updated to: {data}")
            except Exception as e:
                display(str(e))

    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
def delete(messages):
    # Define the form elements
    ruid = widgets.Text(description="Room Unavailable Id:")

    ruid.style.description_width = 'initial'

    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not ruid.value:
                raise Exception("Missing Data!")
            val = int(ruid.value)
            return val
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([ruid])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Delete')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.deleteRoomUnavailable(ruid.value).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Room Unavailable {data['ruid']} deleted!")
            except Exception as e:
                display(str(e))

    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
# Define dropdown options
dropdown_options = ['Select an option', 'Create Room Unavailable']
if data['position'] == "Administrator":
    dropdown_options = ['Select an option', 'Create Room Unavailable', 'Update Room Unavailable', 'Delete Room Unavailable']

# Create the dropdown widget
dropdown = widgets.Dropdown(options=dropdown_options, description='Select Option:', value='Select an option')
output = widgets.Output()
messages = widgets.Output()
vertical_box = widgets.VBox([dropdown, output, messages])

dropdown.style.description_width = 'initial'

# Define a function to handle dropdown changes
def on_dropdown_change(change):
    with output:
        output.clear_output()  # Clear the output area before displaying a new plot
        option = change['new']  # Access the selected option
        if option == "Create Room Unavailable":
            create(messages)
        if option == "Update Room Unavailable":
            update(messages)
        if option == "Delete Room Unavailable":
            delete(messages)
        messages.clear_output()

# Attach the function to the dropdown widget
dropdown.observe(on_dropdown_change, names='value')

# Display the dropdown widget and output widget
display(vertical_box)
```

```python
back_button = widgets.Button(description="Back")

def hide_room_unavailable():
    dropdown.close()
    back_button.close()
    title_1.close()
    
def navigate_index(b):
    with output:
        with open("temp.json", "w") as json_file:
            # Dump the data to the file
            json.dump(data, json_file)
        output.clear_output()
        messages.clear_output()
        hide_room_unavailable()
        %run index.ipynb


back_button.on_click(navigate_index)
display(back_button)
```
