```python
def create(messages):
    # Define the form elements
    dropdown_options_type = ['Select an option', 'Basic', 'Premium', 'Deluxe', 'Suite']
    type = widgets.Dropdown(options=dropdown_options_type, description='Room Type:', value='Select an option')
    type.style.description_width = 'initial'

    dropdown_options_name = ['Select an option', 'Standard', 'Standard Queen', 'Standard King', 'Double Queen', 'Double King', 'Triple King', 'Executive Family', 'Presidential']
    name = widgets.Dropdown(options=dropdown_options_name, description='Room Name:', value='Select an option')
    name.style.description_width = 'initial'

    dropdown_options_capacity = ['Select an option', '1', '2', '4', '6', '8']
    capacity = widgets.Dropdown(options=dropdown_options_capacity, description='Capacity:', value='Select an option')
    capacity.style.description_width = 'initial'

    dropdown_options_handicap = ['Select an option', 'True', 'False']
    handicap = widgets.Dropdown(options=dropdown_options_handicap, description='Handicap:', value='Select an option')
    handicap.style.description_width = 'initial'


    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not name.value or not type.value or not capacity.value or not handicap.value or name.value == "Select an option" or type.value == "Select an option" or capacity.value == "Select an option" or handicap.value == "Select an option":
                raise Exception("Missing Data!")
            values = {
                'rname': name.value,
                'rtype': type.value,
                'capacity': int(capacity.value),
                'ishandicap': handicap.value == "True"
            }
            return values
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([name, type, capacity, handicap])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Create')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.createRoomDescription(json_data).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Room Description created with id: {data['rdid']}")
            except Exception as e:
                display(str(e))

    def on_dropdown_type_change(change):
        if change['type'] == 'change' and change['name'] == 'value':
            type.value = change['new']
    type.observe(on_dropdown_type_change)\
    
    def on_dropdown_name_change(change):
        if change['type'] == 'change' and change['name'] == 'value':
            name.value = change['new']
    name.observe(on_dropdown_name_change)

    def on_dropdown_capacity_change(change):
        if change['type'] == 'change' and change['name'] == 'value':
            capacity.value = change['new']
    capacity.observe(on_dropdown_capacity_change)

    def on_dropdown_handicap_change(change):
        if change['type'] == 'change' and change['name'] == 'value':
            handicap.value = change['new']
    handicap.observe(on_dropdown_handicap_change)


    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
def update(messages):
    # Define the form elements
    rdid = widgets.Text(description='Room Description Id:')
    dropdown_options_type = ['Select an option', 'Basic', 'Premium', 'Deluxe', 'Suite']
    type = widgets.Dropdown(options=dropdown_options_type, description='Room Type:', value='Select an option')

    rdid.style.description_width = 'initial'
    type.style.description_width = 'initial'

    dropdown_options_name = ['Select an option', 'Standard', 'Standard Queen', 'Standard King', 'Double Queen', 'Double King', 'Triple King', 'Executive Family', 'Presidential']
    name = widgets.Dropdown(options=dropdown_options_name, description='Room Name:', value='Select an option')
    name.style.description_width = 'initial'

    dropdown_options_capacity = ['Select an option', '1', '2', '4', '6', '8']
    capacity = widgets.Dropdown(options=dropdown_options_capacity, description='Capacity:', value='Select an option')
    capacity.style.description_width = 'initial'

    dropdown_options_handicap = ['Select an option', 'True', 'False']
    handicap = widgets.Dropdown(options=dropdown_options_handicap, description='Handicap:', value='Select an option')
    handicap.style.description_width = 'initial'

    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not name.value or not type.value or not capacity.value or not handicap.value or name.value == "Select an option" or type.value == "Select an option" or capacity.value == "Select an option" or handicap.value == "Select an option":
                raise Exception("Missing Data!")
            values = {
                'rdid': int(rdid.value),
                'rname': name.value,
                'rtype': type.value,
                'capacity': int(capacity.value),
                'ishandicap': handicap.value == "True"
            }
            return values
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([rdid, name, type, capacity, handicap])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Update')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.updateRoomDescription(json_data, rdid.value).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Employee {data['rdid']} updated to: {data}")
            except Exception as e:
                display(str(e))

    def on_dropdown_type_change(change):
        if change['type'] == 'change' and change['name'] == 'value':
            type.value = change['new']
    type.observe(on_dropdown_type_change)\
    
    def on_dropdown_name_change(change):
        if change['type'] == 'change' and change['name'] == 'value':
            name.value = change['new']
    name.observe(on_dropdown_name_change)

    def on_dropdown_capacity_change(change):
        if change['type'] == 'change' and change['name'] == 'value':
            capacity.value = change['new']
    capacity.observe(on_dropdown_capacity_change)

    def on_dropdown_handicap_change(change):
        if change['type'] == 'change' and change['name'] == 'value':
            handicap.value = change['new']
    handicap.observe(on_dropdown_handicap_change)

    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
def delete(messages):
    # Define the form elements
    rdid = widgets.Text(description="Room Description Id:")
    rdid.style.description_width = 'initial'

    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not rdid.value:
                raise Exception("Missing Data!")
            val = int(rdid.value)
            return val
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([rdid])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Delete')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.deleteRoomDescription(rdid.value).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Employee {data['rdid']} deleted!")
            except Exception as e:
                display(str(e))

    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
# Define dropdown options
dropdown_options = ['Select an option', 'Create Room Description']
if data['position'] == "Administrator":
    dropdown_options = ['Select an option', 'Create Room Description', 'Update Room Description', 'Delete Room Description']

# Create the dropdown widget
dropdown = widgets.Dropdown(options=dropdown_options, description='Select Option:', value='Select an option')
output = widgets.Output()
messages = widgets.Output()
vertical_box = widgets.VBox([dropdown, output, messages])

dropdown.style.description_width = 'initial'

# Define a function to handle dropdown changes
def on_dropdown_change(change):
    with output:
        output.clear_output()  # Clear the output area before displaying a new plot
        option = change['new']  # Access the selected option
        if option == "Create Room Description":
            create(messages)
        if option == "Update Room Description":
            update(messages)
        if option == "Delete Room Description":
            delete(messages)
        messages.clear_output()

# Attach the function to the dropdown widget
dropdown.observe(on_dropdown_change, names='value')

# Display the dropdown widget and output widget
display(vertical_box)
```
