```python
import api.entityCRUD as entityCRUD
import matplotlib.pyplot as plt
import ipywidgets as widgets
import json
```

```python
def create(messages):
    # Define the form elements
    eid = widgets.Text(description='Employee Id:')
    username = widgets.Text(description='Username:')
    password = widgets.Text(description='Password:')

    eid.style.description_width = 'initial'
    username.style.description_width = 'initial'
    password.style.description_width = 'initial'

    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not username.value or not password.value or not eid.value:
                raise Exception("Missing Data!")
            values = {
                'eid': int(eid.value),
                'username': username.value,
                'password': password.value
            }
            return values
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([eid, username, password])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Create')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.createLogin(json_data).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Login created with id: {data['lid']}")
            except Exception as e:
                display(str(e))

    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
def update(messages):
    # Define the form elements
    lid = widgets.Text(description='Login Id:')
    eid = widgets.Text(description='Employee Id:')
    username = widgets.Text(description='Username:')
    password = widgets.Text(description='Password:')

    lid.style.description_width = 'initial'
    eid.style.description_width = 'initial'
    username.style.description_width = 'initial'
    password.style.description_width = 'initial'

    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not username.value or not password.value or not eid.value:
                raise Exception("Missing Data!")
            values = {
                'lid': int(lid.value),
                'eid': int(eid.value),
                'username': username.value,
                'password': password.value
            }
            return values
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([lid, eid, username, password])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Update')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.updateLogin(json_data, lid.value).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Login {data['lid']} updated to: {data}")
            except Exception as e:
                display(str(e))

    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
def delete(messages):
    # Define the form elements
    lid = widgets.Text(description="Login Id:")
    lid.style.description_width = 'initial'

    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not lid.value:
                raise Exception("Missing Data!")
            val = int(lid.value)
            return val
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([lid])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Delete')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.deleteLogin(lid.value).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Login {data['lid']} deleted!")
            except Exception as e:
                display(str(e))

    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
# Define dropdown options
dropdown_options = ['Select an option', 'Create Login', 'Update Login', 'Delete Login']

# Create the dropdown widget
dropdown = widgets.Dropdown(options=dropdown_options, description='Select Option:', value='Select an option')
output = widgets.Output()
messages = widgets.Output()
vertical_box = widgets.VBox([dropdown, output, messages])

dropdown.style.description_width = 'initial'

# Define a function to handle dropdown changes
def on_dropdown_change(change):
    with output:
        output.clear_output()  # Clear the output area before displaying a new plot
        option = change['new']  # Access the selected option
        if option == "Create Login":
            create(messages)
        if option == "Update Login":
            update(messages)
        if option == "Delete Login":
            delete(messages)
        messages.clear_output()

# Attach the function to the dropdown widget
dropdown.observe(on_dropdown_change, names='value')

# Display the dropdown widget and output widget
display(vertical_box)
```
