```python
import api.entityCRUD as entityCRUD
import matplotlib.pyplot as plt
import ipywidgets as widgets
import json
```

```python
def create(messages):
    # Define the form elements
    chid = widgets.Text(description='Chain Id:')
    name = widgets.Text(description='Name:')
    city = widgets.Text(description='City:')


    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not chid.value or not name.value or not city.value:
                raise Exception("Missing Data!")
            values = {
                'chid': int(chid.value),
                'hname': name.value,
                'hcity': city.value
            }
            return values
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([chid, name, city])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Create')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.createHotel(json_data).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Hotel created with id: {data['hid']}")
            except Exception as e:
                display(str(e))

    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
def update(messages):
    # Define the form elements
    hid = widgets.Text(description='Hotel Id:')
    chid = widgets.Text(description='Chain Id:')
    name = widgets.Text(description='Name:')
    city = widgets.Text(description='City:')

    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not hid.value or not chid.value or not name.value or not city.value:
                raise Exception("Missing Data!")
            values = {
                'hid': int(hid.value),
                'chid': int(chid.value),
                'hname': name.value,
                'hcity': city.value
            }
            return values
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([hid, chid, name, city])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Update')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.updateHotel(json_data, hid.value).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Hotel {data['hid']} updated to: {data}")
            except Exception as e:
                display(str(e))

    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
def delete(messages):
    # Define the form elements
    hid = widgets.Text(description="Hotel Id:")

    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not hid.value:
                raise Exception("Missing Data!")
            val = int(hid.value)
            return val
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([hid])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Delete')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.deleteHotel(hid.value).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Hotel {data['hid']} deleted!")
            except Exception as e:
                display(str(e))

    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
# Define dropdown options
dropdown_options = ['Select an option', 'Create Hotel', 'Update Hotel', 'Delete Hotel']

# Create the dropdown widget
dropdown = widgets.Dropdown(options=dropdown_options, description='Select Option:', value='Select an option')
output = widgets.Output()
messages = widgets.Output()
vertical_box = widgets.VBox([dropdown, output, messages])

# Define a function to handle dropdown changes
def on_dropdown_change(change):
    with output:
        output.clear_output()  # Clear the output area before displaying a new plot
        option = change['new']  # Access the selected option
        if option == "Create Hotel":
            create(messages)
        if option == "Update Hotel":
            update(messages)
        if option == "Delete Hotel":
            delete(messages)
        messages.clear_output()

# Attach the function to the dropdown widget
dropdown.observe(on_dropdown_change, names='value')

# Display the dropdown widget and output widget
display(vertical_box)
```
