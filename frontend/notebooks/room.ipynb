```python
import api.entityCRUD as entityCRUD
import matplotlib.pyplot as plt
import ipywidgets as widgets
import json
```

```python
def create(messages):
    # Define the form elements
    hid = widgets.Text(description='Hotel Id:')
    rdid= widgets.Text(description='Room Description Id:')
    rprice = widgets.Text(description='Room Price:')

    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not hid.value or not rdid.value or not rprice.value:
                raise Exception("Missing Data!")
            values = {
                'hid': int(hid.value),
                'rdid': int(rdid.value),
                'rprice': float(rprice.value)
            }
            return values
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([hid, rdid, rprice])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Create')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.createRoom(json_data).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Room created with id: {data['rid']}")
            except Exception as e:
                display(str(e))

    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
def update(messages):
    # Define the form elements
    rid = widgets.Text(description='Room Id:')
    hid = widgets.Text(description='Hotel Id:')
    rdid = widgets.Text(description='Room Description Id:')
    rprice = widgets.Text(description='Room Price:')

    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not rid.value or not hid.value or not hid.value or not rdid.value or not rprice.value:
                raise Exception("Missing Data!")
            values = {
                'rid': int(rid.value),
                'hid': int(hid.value),
                'rdid': int(rdid.value),
                'rprice': float(rprice.value)
            }
            return values
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([rid, hid, rdid, rprice])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Update')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.updateRoom(json_data, rid.value).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Room {data['rid']} updated to: {data}")
            except Exception as e:
                display(str(e))

    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
def delete(messages):
    # Define the form elements
    eid = widgets.Text(description="Employee Id:")

    # Create a function to retrieve values and convert to JSON
    def get_form_values():
        try:
            if not eid.value:
                raise Exception("Missing Data!")
            val = int(eid.value)
            return val
        except Exception as e:
            if str(e) == "Missing Data!":
                return str(e)
            return "Data is not in the correct format!"

    # Display the form
    form = widgets.VBox([eid])
    display(form)

    # Create a button to trigger form submission
    submit_button = widgets.Button(description='Delete')

    # Define a function to handle button click event
    def on_submit_button_clicked(b):
        with messages:
            messages.clear_output()
            json_data = get_form_values()
            try:
                if isinstance(json_data, str):
                    raise Exception(json_data)
                data = entityCRUD.deleteEmployee(eid.value).json()
                if isinstance(data, str):
                    raise Exception(data)
                display(f"Employee {data['eid']} deleted!")
            except Exception as e:
                display(str(e))

    # Attach the click event handler to the button
    submit_button.on_click(on_submit_button_clicked)
    display(submit_button)
```

```python
# Define dropdown options
dropdown_options = ['Select an option', 'Create Room', 'Update Room', 'Delete Room']

# Create the dropdown widget
dropdown = widgets.Dropdown(options=dropdown_options, description='Select Option:', value='Select an option')
output = widgets.Output()
messages = widgets.Output()
vertical_box = widgets.VBox([dropdown, output, messages])

# Define a function to handle dropdown changes
def on_dropdown_change(change):
    with output:
        output.clear_output()  # Clear the output area before displaying a new plot
        option = change['new']  # Access the selected option
        if option == "Create Room":
            create(messages)
        if option == "Update Room":
            update(messages)
        if option == "Delete Room":
            delete(messages)
        messages.clear_output()

# Attach the function to the dropdown widget
dropdown.observe(on_dropdown_change, names='value')

# Display the dropdown widget and output widget
display(vertical_box)
```
